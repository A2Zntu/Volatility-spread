## Define Max Draw Down

```python
def maxdrawdown(arr):
    i = np.argmax((np.maximum.accumulate(arr) - arr)/np.maximum.accumulate(arr)) # end of the period
    print(i)
    j = np.argmax(arr[:i]) # start of period
    print(j)
    return (1-arr[i]/arr[j])
```

## Step1. Construct the Index Buy-and-Hold strategy

```python
start_date = 20070103
end_date = 20171229
list_Date = list(df_SP["Date"])
start_index = list_Date.index(start_date)
end_index = list_Date.index(end_date)
dateList = list(str(date) for date in list_Date)

# The daily return is calculated by log(Pt+1/Pt), therefore returns is accumulated per day
Log_cumReturn = [1, ] # Assume we have one dollar invested at the beginning 
cumReturn = 0 
for i in range(start_index, end_index):
    cumReturn = cumReturn + df_SP["Daily Return"][i]
    Log_cumReturn.append(exp(cumReturn))

freq = 100   
date_length = len(dateList)
date_int = int(date_length//freq)
indexes = list(range(0, date_length, date_int))

cutted_cumR = []
cutted_dateList = []
for ind in indexes:
    cutted_cumR.append(Log_cumReturn[ind])
    cutted_dateList.append(dateList[ind])    
    
plt.figure(figsize=(20,6))
plt.style.use('ggplot')
plt.plot(cutted_dateList, cutted_cumR, '-o', label = 'Investing one dollar at beginning')
plt.xticks(fontsize=8, rotation=45)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Dollars', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
